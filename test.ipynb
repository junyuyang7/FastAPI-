```python
import pandas as pd

data = pd.read_csv(r'C:\Users\Junyu Yang\Desktop\database.csv')
name_id = pd.read_csv(r'C:\Users\Junyu Yang\Desktop\name_id.csv')
```

```python
name = name_id['name'].tolist()
df = pd.merge(data, name_id, left_on='_openid', right_on='openid', how='left')
```

```python
fitter = df[df['name'].isin(name)]
fitter.value_counts()
```

```python
fitter.rename(columns={'qid':'id'}, inplace=True)
fitter.sort_values(by=['name','id'], inplace=True)
fitter.reset_index(drop=True, inplace=True)
```

```python
fitter
```

```python
def compute_acc(df1):
    return round(len(df1[df1['item'] == df1['truth']])/len(df1),3)
acc_data = pd.DataFrame(columns=['acc_d'])
acc_data['acc_d'] = fitter.groupby('name').apply(compute_acc)
fitter = pd.merge(fitter, acc_data, left_on='name', right_on='name', how='left')
```

```python
fitter
```

```python
df_acc = fitter[['name', 'acc_d']].drop_duplicates()
```
